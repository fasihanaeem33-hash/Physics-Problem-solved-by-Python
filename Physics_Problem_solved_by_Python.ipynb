#Problem 1: Calculate Distance Using Speed and Time
Problem Statement: Write a program to calculate the distance traveled by an object, given
its speed and the time it travels.
Formula:
Distance = Speed × Time
Input: Speed (in meters per second) and time (in seconds).
Output: Distance traveled (in meters).
Example:
Input: Speed = 10, Time = 5
Output: Distance = 50 meters

```python
#Calculate Distance Using Speed and Time
# Input from the user
speed = float(input("Enter speed in meters per second: "))
time = float(input("Enter time in seconds: "))

# Calculate distance
distance = speed * time

# Output the result
print("Distance traveled =", distance, "meters")
```

#Problem 2: Calculate Force Using Newton’s Second Law
Problem Statement: Write a program to calculate the force applied on an object using
Newton’s second law of motion.
Formula:
Force = Mass × Acceleration
Input: Mass (in kilograms) and acceleration (in meters per second squared).
Output: Force (in Newtons).
Example:
Input: Mass = 5, Acceleration = 3
Output: Force = 15 N

```python
# Program to calculate force using Newton's second law
# Input from the user
mass = float(input("Enter mass in kilograms: "))
acceleration = float(input("Enter acceleration in m/s^2: "))

# Calculate force
force = mass * acceleration

# Output the result
print("Force =", force, "N")
```

#Problem 3: Gravitational Potential Energy
Problem Statement: Write a program to calculate the gravitational potential energy of an
object.
Formula:
Potential Energy = m × g × h
where g = 9.8 m/s² (acceleration due to gravity), m is the mass, and h is the height.
Input: Mass (in kilograms) and height (in meters).
Output: Gravitational potential energy (in joules).
Example:
Input: Mass = 2, Height = 10
Output: Potential Energy = 196 J

```python
# Program to calculate gravitational potential energy
# Constant for gravity (m/s^2)
g = 9.8

# Input from the user
mass = float(input("Enter mass in kilograms: "))
height = float(input("Enter height in meters: "))

# Calculate potential energy
potential_energy = mass * g * height

# Output the result
print("Potential Energy =", potential_energy, "J")
```

#Problem 4: Kinetic Energy of an Object
Problem Statement: Write a program to calculate the kinetic energy of an object in motion.
Formula:
Kinetic Energy = 1/2 m v²
where m is the mass and v is the velocity.
Input: Mass (in kilograms) and velocity (in meters per second).
Output: Kinetic energy (in joules).
Example:
Input: Mass = 10, Velocity = 3
Output: Kinetic Energy = 45 J

```python
# Program to calculate kinetic energy
# Input from the user
mass = float(input("Enter mass in kilograms: "))
velocity = float(input("Enter velocity in meters per second: "))

# Calculate kinetic energy
kinetic_energy = 0.5 * mass * (velocity ** 2)

# Output the result
print("Kinetic Energy =", kinetic_energy, "J")
```

#Problem 5: Ohm’s Law – Voltage Calculation

current and resistance using Ohm’s Law.

Input: Current (in amperes) and resistance (in ohms).

#Problem 6: Projectile Motion – Maximum Height
Problem Statement: Write a program to calculate the maximum height reached by a
projectile, given its initial velocity and launch angle (ignoring air resistance).
Formula:
h_max = (v² × sin²(θ)) / (2g)
where v is the initial velocity, θ is the launch angle in degrees, and g = 9.8 m/s².
Input: Initial velocity (in meters per second) and angle (in degrees).
Output: Maximum height (in meters).
Example:
Input: Velocity = 20, Angle = 45
Output: Height = 20.4 meters

```python
import math

# Program to calculate maximum height of a projectile
# Constant for gravity
g = 9.8

# Input from the user
velocity = float(input("Enter initial velocity (m/s): "))
angle = float(input("Enter launch angle (degrees): "))

# Convert angle to radians
angle_rad = math.radians(angle)

# Calculate maximum height
h_max = (velocity ** 2) * (math.sin(angle_rad) ** 2) / (2 * g)

# Output the result
print("Maximum Height =", round(h_max, 2), "meters")
```

#Problem 7: Harmonic Oscillator – Period of a Spring
Problem Statement: Write a program to calculate the period of a spring-mass system
(simple harmonic oscillator).
Formula:
T = 2π √(m/k)
where T is the period, m is the mass, and k is the spring constant.
Input: Mass (in kilograms) and spring constant k (in N/m).
Output: Period T (in seconds).
Example:
Input: Mass = 1, Spring constant = 50
Output: Period = 0.89 s

```python
import math

# Program to calculate the period of a spring-mass system
# Input from the user
mass = float(input("Enter mass in kilograms: "))
k = float(input("Enter spring constant (N/m): "))

# Calculate period
T = 2 * math.pi * math.sqrt(mass / k)

# Output the result
print("Period =", round(T, 2), "s")
```

#Problem 8: Free Fall – Time to Hit the Ground
Problem Statement: Write a program to calculate the time it takes for an object to hit the
ground when dropped from a certain height (ignoring air resistance).
Formula:
t = √(2h/g)
where h is the height and g = 9.8 m/s².
Input: Height (in meters).
Output: Time to hit the ground (in seconds).
Example:
Input: Height = 20
Output: Time = 2.02 seconds

```python
import math

# Program to calculate time to hit the ground during free fall
# Constant for gravity (m/s^2)
g = 9.8

# Input from the user
height = float(input("Enter height in meters: "))

# Calculate time
time = math.sqrt((2 * height) / g)

# Output the result
print("Time to hit the ground =", round(time, 2), "seconds")
```

#Problem 9: Work Done by a Force
Problem Statement: Write a program to calculate the work done by a constant force acting
on an object.
Formula:
Work = F × d × cos(θ)
where F is the force, d is the displacement, and θ is the angle between the force and
displacement vectors.
Input: Force (in newtons), displacement (in meters), and angle (in degrees).
Output: Work done (in joules).
Example:
Input: Force = 100, Displacement = 10, Angle = 30
Output: Work = 866 J

```python
import math

# Program to calculate work done by a force
# Input from the user
force = float(input("Enter force in newtons: "))
displacement = float(input("Enter displacement in meters: "))
angle = float(input("Enter angle between force and displacement in degrees: "))

# Convert angle to radians
angle_rad = math.radians(angle)

# Calculate work
work = force * displacement * math.cos(angle_rad)

# Output the result
print("Work =", round(work, 0), "J")
```

#Problem 10: Power Consumption of a Device
Problem Statement: Write a program to calculate the power consumed by an electrical
device.
Formula:
P = W / t
where P is power, W is work (energy), and t is time.
Input: Work (in joules) and time (in seconds).
Output: Power (in watts).
Example:
Input: Work = 500, Time = 50
Output: Power = 10 W

```python
# Program to calculate power consumption
# Input from the user
work = float(input("Enter work done (in joules): "))
time = float(input("Enter time taken (in seconds): "))

# Calculate power
power = work / time

# Output the result
print("Power =", power, "W")
```
